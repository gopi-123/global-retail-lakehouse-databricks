This notebook performs the following tasks to create the Gold Layer Daily Sales table:

1. **Create Gold Layer Daily Sales Table**: Creates the `gold_daily_sales` table in the `globalretail_gold` database.
2. **Aggregate Daily Sales**: Aggregates total sales amounts based on the transaction date from the `silver_orders` table.
3. **Order Data**: Orders the aggregated data based on the transaction date for better readability.

```python
"""
Creates or replaces the 'gold_daily_sales' table in the 'globalretail_gold' database.
Aggregates total sales amounts by transaction date from the 'globalretail_silver.silver_orders' table.
"""

spark.sql("USE globalretail_gold")
spark.sql("""
CREATE OR REPLACE TABLE gold_daily_sales AS
SELECT 
    transaction_date,
    SUM(total_amount) AS daily_total_sales
FROM 
    globalretail_silver.silver_orders
GROUP BY 
    transaction_date
""")
```

```python
%sql
-- Retrieves all records from the gold_daily_sales table, which contains daily aggregated sales totals.
SELECT * FROM gold_daily_sales
```
